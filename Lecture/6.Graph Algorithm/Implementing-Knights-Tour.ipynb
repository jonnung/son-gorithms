# Implementing Knight’s Tour
'기사의 여행' 문제를 해결하기 위해 '**깊이 우선 탐색(Depth First Search, DFS)**'라는 검색 알고리즘을 사용할 것이다.  
**너비 우선 탐색(BFS)**은 검색 트리에서 한개의 레벨(level)씩 빌드를 했다. 깊이 우선 탐색은 검색 트리에서 한개의 가지를 가능한 깊게 탐색한다.  

```python
from lib.pythonds.graphs import Graph, Vertex


def knightTour(n, path, u, limit):
    u.setColor('gray')
    path.append(u)
    if n < limit:
        nbrList = list(u.getConnections())
        i = 0
        done = False
        while i < len(nbrList) and not done:
            if nbrList[i].getColor() == 'white':
                done = knightTour(n+1, path, nbrList[i], limit)
            i = i + 1
        if not done:  # prepare to backtrack
            path.pop()
            u.setColor('white')
    else:
        done = True
    return done
```
